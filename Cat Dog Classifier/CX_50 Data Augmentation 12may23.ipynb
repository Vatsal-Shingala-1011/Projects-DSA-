```python
import tensorflow as tf
import keras.utils as image
from keras.preprocessing.image import ImageDataGenerator
```

```python
#img = image.load_img('''''') not work
#so use
# img = tf.keras.utils.load_img('CampusX_50/cat.PNG',target_size=(200,200))
# or
#replace from keras.preprocessing import image 
# to import keras.utils as image in above shell and use img = image.load_img('''''') not work

img = image.load_img('CampusX_50/cat.PNG',target_size=(200,200))
```

```python
img
```

```python
type(img) #this is not numpy array so can not applied direct numpy operations
```

```python
## convertion img to numpy array
img = image.img_to_array(img) 
type(img)
```

```python
## this object decide that which transformation will be applied
datagen = ImageDataGenerator(
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
img.shape
```

```python
input_batch = img.reshape(1,200,200,3)
```

```python
i=0
    #can write any name instead of output
for output in datagen.flow(input_batch,batch_size=1,save_to_dir='CampusX_50 generated Augumeted images'):
    i=i+1             ## flow for single image                ## store images in this directory
    if i==10:
        break
```
