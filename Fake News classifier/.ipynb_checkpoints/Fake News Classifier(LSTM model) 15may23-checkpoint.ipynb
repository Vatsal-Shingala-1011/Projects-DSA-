```python
###Drop Nan Values
df=df.dropna()
```

```python
## Get the Independent Features
X=df.drop('label',axis=1)
```

```python
## Get the Dependent features
y=df['label']
```

# Onehot Representation

```python
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

```python
### Vocabulary size
voc_size=10000
```

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages.reset_index(inplace=True)#because we have dorp some na values
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
# nltk.download('stopwords')
```

# Dataset Preprocessing

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
#     print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i]) ## substitution all charactors apart from a-a and A-Z with blank
    review = review.lower()
    review = review.split() ##now review = list of words
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
             ##if word in review is not pressent in stopwords.words('english')
    review = ' '.join(review) #joint each array of words with space 
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus] 
onehot_repr
```

# Embedding Representation

```python
sent_length=50
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
embedded_docs[0].shape
```

# Creating model

```python
## Creating model
embedding_vector_features=100 ##dimention
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))#lSTM Layers with 100 neuron
model.add(Dense(1,activation='sigmoid'))#for classificarion problem so for one final output

model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
model.summary()
```

```python
len(embedded_docs) 
```

```python
y.shape
```

```python
import numpy as np
X_final=embedded_docs
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

# Model Training

```python
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred_probs = model.predict(X_test)#;et assume we got probabilities
y_pred = (y_pred_probs > 0.5).astype(int) # Apply thresholding to obtain class predictions

from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score 
accuracy_score(y_test,y_pred)
```
